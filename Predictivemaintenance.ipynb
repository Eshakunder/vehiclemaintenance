```python
X=df_engine.iloc[:,:-1]
Y=df_engine.iloc[:,-1]

print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state = 1, test_size = 0.20)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)
x_train_scaled[0:5]
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled, columns=x_test.columns)
x_train_scaled.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
DTC_model = tree.DecisionTreeClassifier(criterion='entropy',random_state=0)
DTC_model.fit(x_train_scaled, y_train)
 
y_pred_DTC = DTC_model.predict(x_test_scaled)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
RFC_model = RandomForestClassifier(n_estimators=20,criterion="entropy",random_state=0)
RFC_model.fit(x_train_scaled, y_train)
y_pred_RFC = RFC_model.predict(x_test_scaled)

```

```python
from sklearn.svm import SVC   
SVM_model = SVC(kernel='linear', random_state=0)  
SVM_model.fit(x_train_scaled, y_train)  
y_pred_SVM = SVM_model.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
from sklearn import metrics

print("DECISION TREE CLASSIFIER:")
sns.set(rc={"figure.figsize" :(6,6)})
disp=metrics.ConfusionMatrixDisplay.from_predictions(y_test,y_pred_DTC,labels=(1,0),normalize=None,cmap="Blues",)
disp.ax_.grid(False)
print("Accuracy Score : {}".format((accuracy_score(y_test,y_pred_DTC) *100)))
print("Precision Score : {}".format((metrics.precision_score(y_test,y_pred_DTC) *100)))
print("F1 Score : {}".format((metrics.f1_score(y_test,y_pred_DTC) *100)))






```

```python
print("RANDOM FOREST CLASSIFIER:")                                    
sns.set(rc={"figure.figsize" :(6,6)})
disp=metrics.ConfusionMatrixDisplay.from_predictions(y_test,y_pred_RFC,labels=(1,0),normalize=None,cmap="Blues",)
disp.ax_.grid(False)
disp.ax_.set_title("Confusion Matrix :")                                      
print("Accuracy Score : {}".format((accuracy_score(y_test,y_pred_RFC) *100)))
print("Precision Score : {}".format((metrics.precision_score(y_test,y_pred_RFC) *100)))
print("F1 Score : {}".format((metrics.f1_score(y_test,y_pred_RFC) *100)))
```

```python
print("SUPPORT VECTOR MACHINE:")                                    
sns.set(rc={"figure.figsize" :(6,6)})
disp=metrics.ConfusionMatrixDisplay.from_predictions(y_test,y_pred_SVM,labels=(1,0),normalize=None,cmap="Blues",)
disp.ax_.grid(False)
disp.ax_.set_title("SUPPORT VECTOR MACHINE:")                                      
print("Accuracy Score : {}".format((accuracy_score(y_test,y_pred_SVM) *100)))
print("Precision Score : {}".format((metrics.precision_score(y_test,y_pred_SVM) *100)))
print("F1 Score : {}".format((metrics.f1_score(y_test,y_pred_SVM) *100)))
```
